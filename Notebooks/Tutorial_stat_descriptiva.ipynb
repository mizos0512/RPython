# <a id='1'> 1. Introducción</a>

### <a id='1.1'> 1.1  ¿ Qué es la Estadística ?</a>
<div style="text-align: justify;">
La estadística es una forma de análisis matemático que utiliza modelos cuantificados, representaciones y resúmenes para un conjunto dado de datos experimentales o estudios de la vida real. La estadística estudia las metodologías para recopilar, revisar, analizar y sacar conclusiones a partir de los datos.
</div>

<div style="text-align: justify;">
"La estadística significa nunca tener que decir que estás seguro." Si una conclusión estadística alguna vez parece segura, probablemente estés malinterpretando algo. El objetivo de la estadística es cuantificar la incertidumbre.
</div>

### <a id='1.2'> 1.2 Tipos de Estadísticas</a>

<div style="text-align: justify;">

La estadística se utiliza en diversas disciplinas como la psicología, los negocios, las ciencias físicas y sociales, las humanidades, el gobierno y la manufactura. Los datos estadísticos se recopilan utilizando un procedimiento de muestreo u otro método.

Se utilizan dos tipos de métodos estadísticos para analizar los datos:
</div>

 * **Estadística Descriptiva** : 
<div style="text-align: justify;">

La estadística descriptiva se utiliza para resumir datos de una muestra, utilizando medidas como la media o la desviación estándar. La "estadística descriptiva" simplemente describe los datos sin llegar a conclusiones generales. Sin embargo, los aspectos desafiantes y difíciles de la estadística están relacionados con llegar a conclusiones generales a partir de datos limitados.
</div>

 * **Estadística Inferencial** :
<div style="text-align: justify;">

La estadística inferencial se utiliza cuando los datos se consideran como una subclase de una población específica. Con la estadística inferencial, se intenta llegar a conclusiones que vayan más allá de los datos inmediatos. Por ejemplo, utilizamos la estadística inferencial para intentar inferir, a partir de los datos de la muestra, qué podría pensar la población. O utilizamos la estadística inferencial para hacer juicios sobre la probabilidad de que una diferencia observada entre grupos sea confiable o simplemente haya ocurrido por casualidad en este estudio. Por lo tanto, usamos la estadística inferencial para hacer inferencias de nuestros datos a condiciones más generales; utilizamos la estadística descriptiva simplemente para describir lo que está ocurriendo en nuestros datos.
</div>

# <a id='2'> 2. Tipos de Variables: variables Categóricas y Variables Cuantitativas</a>

## <a id='2.1'> 2.1  Variables categóricas </a>

Las variables categóricas clasifican a los individuos o elementos en diferentes grupos.

### Nominal
<div style="text-align: justify;">
Comencemos con la más fácil de entender. Las escalas nominales se utilizan para etiquetar variables, sin asignarles ningún valor cuantitativo. Las escalas "nominales" podrían simplemente llamarse "etiquetas". 
</div>





### Ordinal
<div style="text-align: justify;">

Con las escalas ordinales, el orden de los valores es lo que importa y es significativo, pero las diferencias entre cada uno no son realmente conocidas. Observa el siguiente ejemplo. En cada caso, sabemos que un 4 es mejor que un 3 o un 2, pero no sabemos, ni podemos cuantificar, cuánto mejor es. Por ejemplo, ¿es la diferencia entre "OK" e "Infeliz" la misma que entre "Muy Feliz" y "Feliz"? No podemos afirmarlo.

Las escalas ordinales son típicamente medidas de conceptos no numéricos como la satisfacción, la felicidad, el malestar, etc.
</div>

<div class="alert alert-block alert-info">
<b> Note:</b>   The best way to determine central tendency on a set of ordinal data is to use the mode or median; a purist will tell you that the mean cannot be defined from an ordinal set.
    
</div>

## <a id='2.2'> 2.2  Variables Cuantitativas </a>


Son cantidades numéricas o medibles en las que las operaciones aritméticas a menudo tienen sentido.

### Discreta 

<div style="text-align: justify;">

Contable y finita. Las variables discretas toman un número limitado de valores posibles, como el número de hijos en una familia o el número de autos en un estacionamiento.

</div>

### Continua

<div style="text-align: justify;">

Puede tomar cualquier valor dentro de un intervalo, con muchos posibles valores. Las variables continuas pueden asumir un número infinito de valores en un rango dado, como la altura, el peso o el tiempo.
</div>

## <a id='3.3'>3.3 Datos Cuantitativos - Resumen Numérico</a>

<div style="text-align: justify;">

El resumen de datos proporciona una manera conveniente de describir todos los valores en una muestra de datos con solo unos pocos valores estadísticos.

La media y la desviación estándar se utilizan para resumir datos con una distribución gaussiana, pero pueden no ser significativas o incluso pueden resultar engañosas si tu muestra de datos tiene una distribución no gaussiana.

El resumen de cinco números, o resumen de 5 números para abreviar, es una técnica de resumen de datos no paramétrica.

El resumen de cinco números implica el cálculo de 5 cantidades estadísticas resumen, a saber:

</div>

* Mediana: El valor central en la muestra, también llamado el percentil 50 o el segundo cuartil.

* 1er Cuartil: El percentil 25.
* 3er Cuartil: El percentil 75.
* Mínimo: La observación más pequeña en la muestra.
* Máximo: La observación más grande en la muestra.

## <a id='3.4'>3.4 Puntaje Estándar (Regla Empírica)</a>

<div style="text-align: justify">

El puntaje z (también llamado puntaje estándar) te da una idea de cuán lejos del promedio está un punto de datos. Más técnicamente, es una medida de cuántas desviaciones estándar está un puntaje bruto por debajo o por encima de la media poblacional.

Un puntaje z se puede ubicar en una curva de distribución normal. Los puntajes z van desde -3 desviaciones estándar (lo que estaría en el extremo izquierdo de la curva de distribución normal) hasta +3 desviaciones estándar (lo que estaría en el extremo derecho de la curva de distribución normal). Para usar un puntaje 'z', necesitas conocer la media '$\mu$' y también la desviación estándar de la población '$\sigma$'.

Por ejemplo, saber que el peso de alguien es 150 libras puede ser una buena información, pero si deseas compararlo con el peso "promedio" de una persona, observar una gran tabla de datos puede ser abrumador (especialmente si algunos pesos están registrados en kilogramos). Un puntaje z puede decirte dónde se encuentra el peso de esa persona en comparación con la media del peso de la población promedio.
</div>

### Regla Empírica (68-95-99.7)

<div style="text-align: justify;">

La regla empírica establece que, para una distribución normal, casi todos los datos caerán dentro de tres desviaciones estándar de la media. La regla empírica se puede desglosar en tres partes:

- 68% de los datos cae dentro de la primera desviación estándar desde la media.

- 95% cae dentro de dos desviaciones estándar.

- 99.7% cae dentro de tres desviaciones estándar.

La regla también se conoce como la Regla 68-95-99.7 o la Regla de Tres Sigmas.
</div>



```python
mean = np.mean(titanic_data['Age'])
std = np.std(titanic_data['Age'])
def z_score(value, mean, std):
    return (value - mean) / std
```

```python
import random
values = []
for i in list(range(0,5)):
    value = random.choice(titanic_data.Age)
    values.append(value)
print(values)
```

```python
for val in values:
    z = z_score(val, mean, std)
    print(z)
```

## <a id='3.5'>3.5 Datos Cuantitativos - Diagramas de Cajas</a>

<div style="text-align: justify;">

Un diagrama de cajas (también conocido como diagrama de cajas y bigotes) es un tipo de gráfico a menudo utilizado en el análisis exploratorio de datos para mostrar visualmente la distribución de datos numéricos y la asimetría, mostrando los cuartiles (o percentiles) y promedios de los datos.

Los diagramas de cajas muestran el resumen de cinco números de un conjunto de datos: incluyendo el valor mínimo, el primer cuartil (inferior), la mediana, el tercer cuartil (superior) y el valor máximo.
</div>



**min**

El puntaje más bajo, excluyendo los valores atípicos (mostrados al final del bigote izquierdo).

**lower quartile**

El 25% de los puntajes se encuentra por debajo del valor del cuartil inferior (también conocido como el primer cuartil).

**median**

<div style="text-align: justify;">

La mediana marca el punto medio de los datos y se muestra mediante la línea que divide la caja en dos partes (a veces conocida como el segundo cuartil). La mitad de los puntajes son mayores o iguales a este valor y la otra mitad son menores.
</div>

**upper Quartile**

<div style="text-align: justify;">

El 75% de los puntajes se encuentra por debajo del valor del cuartil superior (también conocido como el tercer cuartil). Por lo tanto, el 25% de los datos están por encima de este valor.
</div>

**max**

El puntaje más alto, excluyendo los valores atípicos (mostrados al final del bigote derecho).

**whisker**

<div style="text-align: justify;">

Los bigotes superior e inferior representan los puntajes fuera del 50% central (es decir, el 25% inferior de los puntajes y el 25% superior de los puntajes).
</div>

**Interquartile Range (IQR)**

Este es el diagrama de cajas que muestra el 50% central de los puntajes (es decir, el rango entre el percentil 25 y el percentil 75).

<div style="text-align: justify;">

>Los diagramas de cajas son útiles ya que proporcionan un resumen visual de los datos, permitiendo a los investigadores identificar rápidamente los valores medios, la dispersión del conjunto de datos y señales de asimetría.
</div>

<div style="text-align: justify;">

>Los diagramas de cajas son útiles ya que muestran los valores atípicos dentro de un conjunto de datos. Un valor atípico es una observación que está numéricamente distante del resto de los datos. Al revisar un diagrama de cajas, un valor atípico se define como un punto de datos que se encuentra fuera de los bigotes del diagrama.
</div>
