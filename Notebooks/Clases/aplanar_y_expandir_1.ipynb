```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'id': [1, 2],
    'data': [
        {'a': [1, 2, 3], 'b': [4, 5]},
        {'a': [6, 7], 'b': [8, 9, 10]}
    ]
}

df = pd.DataFrame(data)
print(df)
#print()

# Paso 1: Expandir la columna 'data' que contiene diccionarios en varias columnas
expanded_df = df['data'].apply(pd.Series)

# Paso 2: Expandir las listas dentro de las columnas 'a' y 'b'
expanded_a = expanded_df['a'].explode()
expanded_b = expanded_df['b'].explode()

# Paso 3: Repetir el índice del DataFrame original para combinar con los datos expandidos
df_repeated = df.loc[df.index.repeat(expanded_df['a'].apply(len))].reset_index(drop=True)
#print(expanded_df['a'].apply(len))
#print()
#print(df.index.repeat(expanded_df['a'].apply(len)))
#print()
#print(df.loc[df.index.repeat(expanded_df['a'].apply(len))])

# Paso 4: Combinar las columnas expandidas con el DataFrame original
result_df = df_repeated.drop(columns=['data']).assign(a=expanded_a.values, b=expanded_b.values)

print(result_df)


"""
    Explicación de:
    df_repeated = df.loc[df.index.repeat(expanded_df['a'].apply(len))].reset_index(drop=True)
    
        data = {
        'id': [1, 2],
        'data': [
           {'a': [1, 2, 3], 'b': [4, 5]},
          {'a': [6, 7], 'b': [8, 9, 10]}
     ]
    }
    
          
     1. expanded_df['a'].apply(len)
     devuelve:
       0    3
       1    2
     porque: la primera columna 'a' es una lista que tiene tres elementos    
             La segunda columna 'a' es una lista que tiene dos elementos
     
     
     2. df.index.repeat(...)
     
     df.index
     Obtiene el índice del DataFrame original 'df' 
     
     repeat(...) 
     Repite cada indice segun el valor corespondiente en la serie resultante del paso anterior   
     
     Es decir según el resultado anterior:
     La instrucción:   
     df.index.repeat(expanded_df['a'].apply(len))
     
     resulta:
     Index([0, 0, 0, 1, 1], dtype='int64')
     
     Es un objeto de tipo pandas.index que es una estructura de datos utilizada por pandas para etiquetar las filas y las columnas de un DataFrame
     El resultado de arriba nos dice: que hay tres indices '0' y dos indices '1' que salieron del resultado anterior
     de expanded_df['a'].apply(len)
     
     3. df.loc[...]
     Selecciona las filas del DataFrame 'df' basandose en los indices repetidos generados en el paso anterior:
     df.loc[df.index.repeat(expanded_df['a'].apply(len))]
     id                            data
    0   1   {'a': [1, 2, 3], 'b': [4, 5]}
    0   1   {'a': [1, 2, 3], 'b': [4, 5]}
    0   1   {'a': [1, 2, 3], 'b': [4, 5]}
    1   2  {'a': [6, 7], 'b': [8, 9, 10]}
    1   2  {'a': [6, 7], 'b': [8, 9, 10]}
             
             
             
    
    
"""
```
