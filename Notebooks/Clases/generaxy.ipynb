```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# Configuración inicial
np.random.seed(0)
n_samples = 100  # Número de muestras

# Generar datos para las variables independientes
X = np.abs(np.random.randn(n_samples, 5) * 100)  # Asegurarse de que X no tenga valores negativos

# Coeficientes del modelo
true_coefficients = np.array([0.3, -0.2, 0.5, 0.1, -0.4])
intercept = 50

# Generar errores
# Estimación inicial de la varianza total
variance_true = np.var(X @ true_coefficients + intercept)
target_r2 = 0.85
variance_error = variance_true * (1 - target_r2) / target_r2
error_std = np.sqrt(variance_error)
errors = np.random.randn(n_samples) * error_std

# Generar la variable dependiente, asegurándonos de que sea no negativa
Y = X @ true_coefficients + intercept + errors
Y = np.maximum(Y, 0)  # Forzar que todos los valores de Y sean no negativos

# Crear un modelo de regresión y ajustarlo
model = LinearRegression()
model.fit(X, Y)

# Calcular R-squared con el modelo ajustado
Y_pred = model.predict(X)
r_squared = r2_score(Y, Y_pred)

print(f"R-squared: {r_squared:.3f}")

# Crear y guardar el DataFrame
df = pd.DataFrame(X, columns=[f'X{i+1}' for i in range(5)])
df['Y'] = Y
nombreArchivo = "C:/Users/CursosTardes/datos_multivariables.csv"
df.to_csv(nombreArchivo, index=False)

# Mostrar los primeros datos
print(df.head())

# Plotear los datos
fig = plt.figure(figsize=(18, 6))

# Graficar un par de variables independientes contra Y
ax1 = fig.add_subplot(131)
ax1.scatter(X[:, 0], Y, alpha=0.7, label='Datos')
ax1.set_xlabel('X1')
ax1.set_ylabel('Y')
ax1.set_title('X1 vs Y')
ax1.legend()

ax2 = fig.add_subplot(132)
ax2.scatter(X[:, 1], Y, alpha=0.7, label='Datos')
ax2.set_xlabel('X2')
ax2.set_ylabel('Y')
ax2.set_title('X2 vs Y')
ax2.legend()

# Graficar en 3D las primeras dos variables independientes
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(X[:, 0], X[:, 1], Y, c='r', marker='o')
ax3.set_xlabel('X1')
ax3.set_ylabel('X2')
ax3.set_zlabel('Y')
ax3.set_title('X1 vs X2 vs Y')

plt.tight_layout()
plt.show()


```
