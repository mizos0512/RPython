### **Sesgos en Estadística y Aprendizaje Automático**

**1. Sesgo (Bias) en Estadística:**

En estadística, el sesgo se refiere a la tendencia sistemática de un estimador para alejarse del valor verdadero del parámetro que se está estimando. Un estimador sesgado no necesariamente se aproxima al valor real del parámetro en promedio.

- **Sesgo de Estimador:**
  - **Definición:** Diferencia entre el valor esperado del estimador y el valor verdadero del parámetro que se está estimando.
  - **Ejemplo:** Si utilizas una muestra para estimar la media de una población, el sesgo es la diferencia entre la media de la muestra y la media verdadera de la población.

**2. Sesgo (Bias) en Aprendizaje Automático:**

En el contexto de aprendizaje automático, el sesgo puede tener varios significados dependiendo del contexto:

- **Sesgo del Modelo (Bias):**
  - **Definición:** La capacidad de un modelo para hacer supuestos incorrectos sobre los datos. Un alto sesgo generalmente resulta en un modelo subajustado (underfitting), que no captura adecuadamente la complejidad de los datos.
  - **Ejemplo:** Un modelo lineal que intenta ajustar datos con una relación no lineal puede tener un alto sesgo, ya que no puede capturar la complejidad de la relación.

- **Sesgo de Datos (Data Bias):**
  - **Definición:** Ocurre cuando los datos de entrenamiento no son representativos de la población general. Esto puede llevar a un modelo que funciona bien en el conjunto de datos de entrenamiento pero no en datos nuevos.
  - **Ejemplo:** Si un modelo de reconocimiento facial se entrena principalmente con imágenes de personas de una raza específica, puede tener un sesgo hacia esa raza y funcionar mal con personas de otras razas.

- **Sesgo de Muestreo:**
  - **Definición:** Se produce cuando el proceso de muestreo introduce una selección sesgada, lo que puede llevar a conclusiones incorrectas.
  - **Ejemplo:** Si solo se recopilan datos de un grupo demográfico específico, el modelo puede no generalizar bien a otros grupos.

- **Sesgo de Confirmación:**
  - **Definición:** Tendencia a buscar, interpretar o recordar información de manera que confirme las propias creencias preexistentes.
  - **Ejemplo:** Un científico que tiene una hipótesis específica puede buscar solo datos que apoyen esa hipótesis, ignorando datos que podrían contradecirla.

**3. Compensación entre Sesgo y Varianza:**

En aprendizaje automático, se suele hablar de la **compensación entre sesgo y varianza** (Bias-Variance Tradeoff):

- **Sesgo Bajo y Varianza Alta:** Un modelo con sesgo bajo y varianza alta puede ajustarse demasiado a los datos de entrenamiento (overfitting), capturando ruido en lugar de patrones subyacentes.
- **Sesgo Alto y Varianza Baja:** Un modelo con sesgo alto y varianza baja puede no ajustarse bien a los datos de entrenamiento (underfitting), perdiendo la capacidad de capturar patrones importantes.

**Resumen:**

- **Sesgo en Estadística:** Diferencia entre el valor esperado del estimador y el valor verdadero del parámetro.
- **Sesgo en Aprendizaje Automático:**
  - **Sesgo del Modelo:** Tendencia del modelo a hacer suposiciones incorrectas.
  - **Sesgo de Datos:** Datos de entrenamiento no representativos.
  - **Sesgo de Muestreo:** Introducción de sesgo en el proceso de muestreo.
  - **Sesgo de Confirmación:** Interpretación sesgada de la información para confirmar creencias preexistentes.
- **Compensación Sesgo-Varianza:** Equilibrio entre un modelo que es demasiado rígido (alto sesgo) y uno que se ajusta demasiado a los datos (alta varianza).

### **Influencia del Sesgo en Machine Learning**

**1. ¿Qué es el Sesgo en Machine Learning?**

En el contexto del aprendizaje automático, el sesgo se refiere a la tendencia de un modelo para hacer suposiciones simplificadas sobre el conjunto de datos, lo cual puede llevar a errores sistemáticos en las predicciones. Un alto sesgo puede resultar en un modelo que no captura adecuadamente las complejidades de los datos, llevando a un problema de subajuste (underfitting).

**2. ¿Cómo Afecta el Sesgo al Rendimiento del Modelo?**

- **Subajuste (Underfitting):** Cuando un modelo tiene un sesgo alto, tiende a simplificar demasiado la relación entre las características y la variable objetivo, lo que resulta en un rendimiento deficiente tanto en los datos de entrenamiento como en los datos de prueba. El modelo no captura patrones importantes en los datos, y su capacidad predictiva es limitada.
  
- **Generalización Pobre:** Un modelo con alto sesgo no generaliza bien a nuevos datos. Esto ocurre porque el modelo es demasiado rígido y no puede adaptarse a las variaciones presentes en el conjunto de datos.

**3. Caso Breve: Regresión Lineal para Predicción de Precios de Viviendas**

Imaginemos que estamos utilizando un modelo de regresión lineal para predecir los precios de viviendas en una ciudad. Nuestra característica principal es el tamaño de la vivienda en metros cuadrados.

- **Modelo con Alto Sesgo:**
  - **Descripción:** Usamos una regresión lineal simple que solo considera el tamaño de la vivienda como predictor del precio.
  - **Problema:** El modelo asume que el precio de la vivienda solo depende linealmente del tamaño. Sin embargo, el precio también puede depender de otros factores importantes, como la ubicación, el número de habitaciones, el año de construcción, etc.
  - **Impacto:** El modelo de regresión lineal con alto sesgo no captura las complejidades adicionales (como la ubicación) y, por lo tanto, tiene un rendimiento deficiente. Los errores de predicción son grandes tanto para datos de entrenamiento como para datos nuevos.

**4. Ejemplo en Código:**

Para ilustrar cómo el sesgo puede afectar a un modelo, a continuación se muestra un ejemplo en Python usando `scikit-learn` para una regresión lineal simple.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Generar datos sintéticos
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Tamaño de la vivienda
y = 2.5 * X.squeeze() + np.random.randn(100) * 2  # Precio de la vivienda con algo de ruido

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Crear y ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
print(f"Error Cuadrático Medio (MSE): {mse}")

# Visualizar resultados
plt.scatter(X_test, y_test, color='blue', label='Datos de Prueba')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Ajuste del Modelo')
plt.xlabel('Tamaño de la Vivienda (m²)')
plt.ylabel('Precio de la Vivienda')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
````

**5. Conclusión:**

El sesgo alto en un modelo de aprendizaje automático puede llevar a una pobre capacidad de ajuste y generalización. En el caso de la regresión lineal simple, al no considerar múltiples factores que afectan el precio de las viviendas, el modelo resultante tiene un rendimiento limitado. Ajustar el sesgo, a menudo mediante el uso de modelos más complejos o la inclusión de características adicionales, es crucial para mejorar el rendimiento del modelo y la precisión de las predicciones.

### a. Observación en el gráfico en términos de regresión a la media:

En el gráfico, podemos observar que los estudiantes que obtuvieron puntuaciones extremadamente altas en el primer examen tienden a tener puntuaciones más bajas en el segundo examen, mientras que aquellos que obtuvieron puntuaciones extremadamente bajas en el primer examen tienden a mejorar sus puntuaciones en el segundo examen. En otras palabras, las puntuaciones extremas en el primer examen tienden a acercarse al promedio en el segundo examen. Esta tendencia ilustra el fenómeno de la **regresión a la media**.

### b. Razón de la tendencia hacia el promedio:

La regresión a la media ocurre porque los resultados extremadamente altos o bajos en el primer examen pueden estar influidos por factores aleatorios o específicos que no necesariamente se repetirán en el segundo examen. Por ejemplo, un estudiante que tuvo un desempeño extremadamente bueno en el primer examen podría haber sido favorecido por una combinación de suerte o preguntas que se alineaban perfectamente con sus conocimientos. Sin embargo, es probable que en un segundo examen, la influencia de esos factores aleatorios se diluya, y su puntuación se acerque más a su verdadero promedio. De manera similar, un estudiante que tuvo un desempeño muy bajo en el primer examen podría haber sido desfavorecido por la suerte o factores momentáneos (como estrés o falta de preparación), y en el segundo examen, su puntuación tiende a mejorar y acercarse a la media.

### c. Aplicación en otras áreas:

El concepto de regresión a la media se puede aplicar en diversas áreas, como:

1. **Deportes**: En el análisis de rendimiento deportivo, un atleta que tiene un rendimiento excepcionalmente bueno o malo en una competencia puede tener un rendimiento más cercano a su promedio en la siguiente competencia. Esto es importante para entrenadores y analistas que no deberían sobreinterpretar un rendimiento excepcional como una nueva norma, sino esperar una regresión hacia el promedio en futuras actuaciones.

2. **Predicción de ventas**: En el ámbito de las ventas, si un producto tiene un pico de ventas excepcional en un periodo determinado (por ejemplo, debido a una promoción especial o un evento), es probable que las ventas regresen a un nivel más cercano al promedio en periodos futuros, una vez que esos factores excepcionales desaparezcan.

Este concepto es crucial para no hacer proyecciones o ajustes excesivos basados en resultados extremos que pueden ser en gran parte debido al azar o circunstancias temporales.

````python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=500)  # Media = 50, Desviación estándar = 10, Tamaño de la muestra = 500

mean = np.mean(data)

sem = stats.sem(data)  

confidence_interval = stats.t.interval(0.98, len(data)-1, loc=mean, scale=sem)

ci_lower = float(confidence_interval[0])
ci_upper = float(confidence_interval[1])

print(f"Media de la muestra: {mean:.2f}")
print(f"Intervalo de confianza del 98%: ({ci_lower:.2f}, {ci_upper:.2f})")
````

Este código calcula un **intervalo de confianza del 98%** para la media de una muestra de datos generados a partir de una distribución normal. 

### Explicación del código

1. **`import numpy as np` y `import scipy.stats as stats`**:
   - Importa las librerías necesarias. `numpy` (abreviado como `np`) se utiliza para generar los datos y calcular la media, mientras que `scipy.stats` (abreviado como `stats`) se utiliza para calcular el error estándar de la media y el intervalo de confianza.

2. **`np.random.seed(42)`**:
   - Fija la semilla del generador de números aleatorios en 42. Esto asegura que los resultados sean reproducibles, es decir, que cada vez que ejecutes el código, se generarán los mismos números aleatorios.

3. **`data = np.random.normal(loc=50, scale=10, size=500)`**:
   - Genera una muestra de datos aleatorios que sigue una distribución **normal**. 
     - `loc=50`: Es la media de la distribución normal.
     - `scale=10`: Es la desviación estándar de la distribución normal.
     - `size=500`: Es el tamaño de la muestra (500 datos aleatorios).
   - Por lo tanto, se genera una muestra de 500 datos con media 50 y desviación estándar 10.

4. **`mean = np.mean(data)`**:
   - Calcula la **media de la muestra** `data` usando la función `mean()` de `numpy`.

5. **`sem = stats.sem(data)`**:
   - Calcula el **error estándar de la media** (SEM, por sus siglas en inglés), que es la desviación estándar de la media muestral. `stats.sem()` toma como entrada los datos de la muestra y calcula el error estándar, que es una medida de la precisión de la media muestral.

6. **`confidence_interval = stats.t.interval(0.98, len(data)-1, loc=mean, scale=sem)`**:
   - Calcula el **intervalo de confianza del 98%** para la media de la muestra usando una distribución t de Student.
     - `0.98`: Nivel de confianza del 98%.
     - `len(data)-1`: Los grados de libertad, que son el tamaño de la muestra menos 1.
     - `loc=mean`: Es la media muestral.
     - `scale=sem`: Es el error estándar de la media.
   - El resultado es una tupla con los límites inferior y superior del intervalo de confianza.

7. **`ci_lower = float(confidence_interval[0])` y `ci_upper = float(confidence_interval[1])`**:
   - Extrae el límite inferior (`ci_lower`) y el límite superior (`ci_upper`) del intervalo de confianza y los convierte en valores de punto flotante para presentarlos con formato.

8. **`print(f"Media de la muestra: {mean:.2f}")`**:
   - Imprime la media muestral con dos decimales.

9. **`print(f"Intervalo de confianza del 98%: ({ci_lower:.2f}, {ci_upper:.2f})")`**:
   - Imprime el intervalo de confianza del 98%, con los límites inferior y superior redondeados a dos decimales.

### Ejemplo de resultado

```
Media de la muestra: 49.71
Intervalo de confianza del 98%: (48.34, 51.09)
```

### Interpretación

1. **Media de la muestra**: El valor promedio de los datos generados es aproximadamente 49.71. Esto es cercano a la media real de la distribución que generó los datos (50).

2. **Intervalo de confianza del 98%**: 
   - El intervalo de confianza del 98% es (48.34, 51.09). Esto significa que, si se repitieran muchos experimentos similares y se calcularan los intervalos de confianza para cada uno, el 98% de esos intervalos contendría la media verdadera (50 en este caso). 
   - En otras palabras, hay un 98% de probabilidad de que la media verdadera de la población esté dentro de este intervalo.

**Aplicaciones y casos de uso del Bootstrapping en Machine Learning**

**Bootstrapping** es una técnica estadística que se utiliza para generar múltiples muestras a partir de una muestra de datos original, mediante la selección aleatoria con reemplazo. En Machine Learning, esta técnica tiene diversas aplicaciones, especialmente cuando se busca mejorar la estimación de modelos, validar resultados, o trabajar con datos limitados.

### Principales aplicaciones:

1. **Estimación de incertidumbre y varianza**:
   - En Machine Learning, uno de los retos es estimar la incertidumbre en los modelos predictivos. El bootstrapping permite calcular métricas como la media, desviación estándar y los intervalos de confianza para las predicciones del modelo.
   - Este enfoque es útil cuando no se puede asumir una distribución teórica para los errores, lo que hace que el bootstrapping sea un método robusto y no paramétrico.

2. **Validación cruzada y evaluación de modelos**:
   - Se emplea en la evaluación de modelos predictivos para obtener un mejor estimador de su rendimiento. Al crear múltiples conjuntos de datos de entrenamiento y prueba mediante bootstrapping, se puede calcular un promedio más fiable de las métricas de rendimiento (por ejemplo, precisión, F1-score).
   - La técnica **out-of-bag** (OOB) es común en modelos como Random Forest, donde cada árbol se entrena con un subconjunto bootstrap del conjunto de datos original, y los datos que no se seleccionan se utilizan para evaluar la precisión de las predicciones del modelo.

3. **Bagging (Bootstrap Aggregating)**:
   La agregación de bootstrap o autodocimante, conocida por su nombre en inglés bagging, es el método de aprendizaje por conjuntos que se suele utilizar para reducir la varianza dentro de un conjunto de datos ruidoso. En el bagging, se selecciona una muestra aleatoria de datos en un conjunto de entrenamiento con reemplazo, lo que significa que los puntos de datos individuales se pueden elegir más de una vez. Tras generar varias muestras de datos, estos modelos débiles se entrenan de forma independiente, y en función del tipo de tarea —regresión o clasificación, por ejemplo— el promedio o la mayoría de estas predicciones genera una estimación más precisa. 

   Como nota, el algoritmo de bosque aleatorio se considera una extensión de bagging, ya que utiliza tanto este método como la aleatoriedad de características para crear un bosque no correlacionado de árboles de decisiones.

4. **Generación de ensambles**:
   - El ensamble de modelos combina múltiples algoritmos para mejorar el rendimiento general. El bootstrapping es una herramienta clave en la creación de estos ensambles, ya que proporciona variabilidad al crear diferentes subconjuntos de datos de entrenamiento para cada modelo.
   - Este enfoque no solo mejora la precisión, sino que también es útil para manejar la incertidumbre de las predicciones en dominios como el reconocimiento de imágenes, predicción de series temporales, y procesamiento de lenguaje natural.

5. **Imputación de datos faltantes**:
   - Cuando hay datos faltantes en un conjunto, el bootstrapping se puede utilizar para crear diferentes imputaciones o estimaciones de los valores ausentes. Se generan múltiples conjuntos de datos con diferentes imputaciones que se promedian para proporcionar una mejor estimación final, lo cual es útil en problemas de datos incompletos.

### Casos de uso:

- **Evaluación en finanzas**: En el ámbito de las finanzas, el bootstrapping es un enfoque metodológico para estimar la curva de rendimiento a partir de los precios de un conjunto de productos con cupones. - **Reconocimiento de imágenes**: Al aplicar bagging (Bootstrap Aggregating), se entrenan múltiples modelos de reconocimiento de imágenes sobre diferentes subconjuntos del conjunto de datos, creados mediante el método de bootstrap. Estos modelos pueden ser redes neuronales, árboles de decisión o cualquier otro tipo de modelo de alta varianza. Posteriormente, se promedian las predicciones (en el caso de problemas de regresión) o se utiliza un voto mayoritario (en clasificación) para generar una predicción final más estable y precisa.

En Resumen, Bootstrap es una técnica estadística no paramétrica que genera múltiples subconjuntos de datos mediante la selección aleatoria con reemplazo. En Machine Learning, tiene un papel crucial en mejorar la precisión, estabilidad y generalización de los modelos. Su versatilidad permite aplicarlo en diferentes aspectos del entrenamiento y evaluación de modelos.

## Intervalos basados en percentiles


Los intervalos basados en percentiles se basan en la idea de que, la distribución simulada, se asemeja a la distribución muestral del estadístico, y de que su desviación estándar se aproxima al error estándar de dicha distribución. Acorde a esto, se puede estimar el intervalo de confianza del estadístico utilizando los percentiles de la distribución obtenida.
### Ejemplo

Se dispone de una muestra formada por 30 observaciones de una variable aleatoria continua. Se desea calcular un intervalo de confianza del 95% para la media empleando el método de bootstrapping basado en percentiles.

La representación gráfica muestra evidencias de que los datos no se distribuyen de forma normal. Esto implica que, la aproximación basada en el teorema del límite central para estimar el error estándar  SE=sd/√n
 , deja de ser buena y con ella los intervalos paramétricos basados en la estructura  [parámetro estimado ± tα SE]
 . Una alternativa para poder calcular intervalos de confianza es emplear bootstrapping.

### Intervalo de confianza basado en percentiles

Mediante bootstrapping, se simula la variabilidad esperada en el estadístico, en este caso la media, debido únicamente al muestreo aleatorio.

````python
def calcular_estadistico(x):
    '''
    Función para calcular el estadístico de interés.
    
    Parameters
    ----------
    x : numpy array
         valores de la muestra.
         
    Returns
    -------
    estadístico: float
        valor del estadístico.
    '''
    estadistico = np.mean(x)
    
    return(estadistico)


def bootstraping(x, fun_estadistico, n_iteraciones=9999):
    '''
    Función para calcular el valor del estadístico en múltiples muestras generadas
    mediante muestreo repetido con reposición (bootstrapping).
    
    Parameters
    ----------
    x : numpy array
         valores de la muestra.
 
    fun_estadistico : function
        función que recibe como argumento una muestra y devuelve el valor
        del estadístico.
        
    n_iteraciones : int
        número iteraciones (default `9999`).
        
    Returns
    -------
    distribuciones: numpy array
        valor del estadístico en cada muestra de bootstrapping.
    '''
    
    n = len(x)
    dist_boot = np.full(shape=n_iteraciones, fill_value=np.nan)
    
    for i in tqdm(range(n_iteraciones)):
        resample = np.random.choice(x, size=n, replace=True)
        dist_boot[i] = fun_estadistico(resample)
        
    return dist_boot
````

````python
dist_boot = bootstraping(
                x = datos,
                fun_estadistico = calcular_estadistico,
                n_iteraciones   = 9999
            )
````

````python
# Distribución de bootstrapping
# ==============================================================================
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(7,3.3))
ax.hist(dist_boot, bins=30, density=True, color='#3182bd', alpha=0.5)
ax.set_title('Distribución de bootstrapping')
ax.set_xlabel('media')
ax.set_ylabel('densidad');
````

La dispersión de la distribución obtenida por bootstrapping es una aproximación del error estándar esperado debido a proceso de muestreo. Por esta razón, pueden emplearse sus percentiles para calcular intervalos de confianza.

````python
# Intervalo IC basado en percentiles de la distribución bootstrapping
# ==============================================================================
# Un IC del 95% debe abarcar desde el cuantil 0.025 al 0.975
cuantiles = np.quantile(a = dist_boot, q = [0.025, 0.975])
print('-------------------------------')
print('Intervalo basado en percentiles')
print('-------------------------------')
print(cuantiles)
````

````python
# Gráfico intervalo de confianza del 95%
# ==============================================================================
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(7,4))
ax.hist(dist_boot, bins=30, density=True, color='#3182bd', alpha=0.5)
ax.axvline(x=datos.mean(), color='firebrick', label='media observada')
ax.axvline(x=cuantiles[0], color='black', linestyle='--', label='IC 95%')
ax.axvline(x=cuantiles[1], color='black', linestyle='--')
ax.hlines(y=0.001, xmin=cuantiles[0], xmax=cuantiles[1], color='black')
ax.set_title('Intervalo bootstrapping basados en percentiles')
ax.set_xlabel('media')
ax.set_ylabel('densidad')
ax.legend();
````

#### Intervalos de confianza empírico

A diferencia de las otras aproximaciones, en el bootstrapping empírico, la distribución simulada no es la del estadístico de interés (la media en este caso), sino la de las desviaciones entre el estadístico calculado con la muestra original y el estadístico obtenido en cada muestra bootstrap.

````python
def calcular_estadistico(x):
    '''
    Función para calcular el estadístico de interés.
    
    Parameters
    ----------
    x : numpy array
         valores de la muestra.
         
    Returns
    -------
    estadístico: float
        valor del estadístico.
    '''
    estadistico = np.mean(x)
    
    return(estadistico)


def bootstraping_empirico(x, fun_estadistico, n_iteraciones=9999):
    '''
    Función para calcular la diferencia entre el valor del estadístico en la
    muestra original y el valor del estadístico en múltiples muestras generadas
    por muestreo con reposición.
    
    Parameters
    ----------
    x : numpy array
         valores de la muestra.
 
    fun_estadistico : function
        función que recibe como argumento una muestra y devuelve el valor
        del estadístico.
        
    n_iteraciones : int
        número iteraciones (default `9999`).
        
    Returns
    -------
    distribuciones: numpy array
        valor de la diferencia del estadístico en cada muestra de bootstrapping.
    '''
    
    n = len(x)
    estadistico_muestra = fun_estadistico(x)
    dist_boot = np.full(shape=n_iteraciones, fill_value=np.nan)
    
    for i in tqdm(range(n_iteraciones)):
        resample = np.random.choice(x, size=n, replace=True)
        dist_boot[i] = fun_estadistico(resample) - estadistico_muestra
        
    return dist_boot
````

````python
dist_boot = bootstraping_empirico(
                x = datos,
                fun_estadistico = calcular_estadistico,
                n_iteraciones   = 9999
            )
````

````python
# Distribución de bootstrapping
# ==============================================================================
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(7,3.3))
ax.hist(dist_boot, bins=30, density=True, color='#3182bd', alpha=0.5)
ax.set_title('Distribución de bootstrapping')
ax.set_xlabel(r'$\hat{X}^{b} - \hat{X}$')
ax.set_ylabel('densidad');
````

La distribución obtenida por bootstrapping es una aproximación de la diferencia que cabe esperar entre el valor del estadístico calculado con los datos de la muestra disponible y el verdadero valor del parámetro poblacional. En este ejemplo en concreto, representa la diferencia que se puede esperar entre la media de la muestra y el valor medio de la población.

````python
# Intervalo IC basado en bootstrapping empírico.
# ==============================================================================
# Un IC del 95% debe abarcar desde el cuantil 0.025 al 0.975
cuantiles = np.quantile(a = dist_boot, q = [0.025, 0.975])
estadistico_muestra = calcular_estadistico(datos)
print('------------------')
print('Intervalo empírico')
print('------------------')
intervalo = [estadistico_muestra - cuantiles[1] , estadistico_muestra - cuantiles[0]]
print(intervalo)
````

### Distribución bootstrapping

El objetivo del estudio es determinar si, la diferencia observada de 5.9 unidades, está dentro de lo que cabría esperar debido únicamente al muestreo aleatorio, si no existiera diferencia real entre las poblaciones ( H0
 : todas las observaciones proceden de la misma población).

Si se obtuviera una nueva muestra de estudiantes, la diferencia promedio entre grupos sería ligeramente distinta, aunque no existiera una diferencia real entre nacionalidades. Por lo tanto, dar respuesta a este problema pasa por determinar cuánta diferencia se espera por el simple hecho de repetir el muestreo.

Mediante bootstrapping, se generan nuevas pseudo-muestras de individuos empleando las dos muestras originales (país 1 y país 2) combinadas. Con cada una de las pseudo-muestras, se generan dos grupos de tamaños iguales a los grupos de las muestras originales (34, 116) y se calcula la diferencia del estadístico, en este caso la media. El proceso se asemeja mucho al empleado en los test de permutación. La diferencia radica en que, en los test de permutación, se emplean siempre las observaciones de la muestra original ordenadas de forma distinta en cada iteración. En este caso, en cada iteración y antes del reparto en grupos, se genera una nueva pseudo-muestra que tiene el mismo tamaño que la muestra original pero formada por distintas observaciones.

````python
def calcular_estadistico(x):
    '''
    Función para calcular el estadístico de interés.
    
    Parameters
    ----------
    x : numpy array
         valores de la muestra.
         
    Returns
    -------
    estadístico: float
        valor del estadístico.
    '''
    estadistico = np.mean(x)
    
    return(estadistico)


def bootstraping_2_grupos(x1, x2, fun_estadistico, n_iteraciones=9999):
    '''
    Función para calcular la diferencia de un estadístico entre dos grupos en
    múltiples muestras generadas mediante muestreo repetido con reposición
    (bootstrapping).
    
    Parameters
    ----------
    x1 : numpy array
         valores de la muestra del grupo 1.
         
    x2 : numpy array
         valores de la muestra del grupo 2.
 
    fun_estadistico : function
        función que recibe como argumento una muestra y devuelve el valor
        del estadístico.
        
    n_iteraciones : int
        número iteraciones (default `9999`).
        
    Returns
    -------
    distribuciones: numpy array
        diferencia entre ambos grupos en cada muestra de bootstrapping.
    '''
    
    n1 = len(x1)
    n2 = len(x2)
    pool = np.hstack((x1, x2))
    dist_boot = np.full(shape=n_iteraciones, fill_value=np.nan)
    
    for i in tqdm(range(n_iteraciones)):
        # Se crea una nueva muestra
        resample = np.random.choice(pool, size=n1+n2, replace=True)
        # Se reparten las observaciones en dos grupos y se calcula el estadístico
        estadistico_1 = fun_estadistico(resample[:n1])
        estadistico_2 = fun_estadistico(resample[n1:])
        # Diferencia entre estadísticos
        dist_boot[i]  = estadistico_1 - estadistico_2
        
    return dist_boot
````

````python
dist_boot = bootstraping_2_grupos(
                x1 = datos.nota[datos.nacionalidad == 'pais_1'],
                x2 = datos.nota[datos.nacionalidad == 'pais_2'],
                fun_estadistico = calcular_estadistico,
                n_iteraciones   = 9999
            )
````

Los datos simulados forman lo que se conoce como bootstrapping distribution y representa la variación esperada en la diferencia de medias debida únicamente al muestreo aleatorio, si ambas muestras proceden de la misma población.

````python
# Distribución de bootstrapping
# ==============================================================================
dif_observada = datos.nota[datos.nacionalidad == 'pais_1'].mean() \
                - datos.nota[datos.nacionalidad == 'pais_2'].mean()

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(7,4))
ax.hist(dist_boot, bins=30, density=True, color='#3182bd', alpha=0.5)
ax.axvline(x=dif_observada, color='red', label='diferencia observada')
ax.axvline(x=-dif_observada, color='red')

ax.set_title('Distribución de bootstrapping')
ax.set_xlabel('diferencia de medias')
ax.set_ylabel('densidad')
ax.legend();
````

````python
pd.Series(dist_boot).describe()
````

Como era de esperar, la diferencia entre la media de los grupos, considerando que todos proceden de una misma población, está centrada en cero.

#### p-value

Finalmente, se calcula la probabilidad (p-value) de obtener diferencias iguales o más extremas que la observada (líneas verticales rojas).

````python
# P-value empírico con y sin corrección
# ==============================================================================
p_value = (sum(np.abs(dist_boot) > np.abs(dif_observada)))/len(dist_boot)
p_value_correc = (sum(np.abs(dist_boot) > np.abs(dif_observada)) + 1)/len(dist_boot + 1)
print(f"p-value sin corrección: {p_value}")
print(f"p-value con corrección: {p_value_correc}")
````

### Conclusión

Un test de bootstrapping no paramétrico se empleó para determinar si existe diferencia en el promedio de logros académicos entre estudiantes procedentes de dos nacionalidades distintas. Los datos se obtuvieron mediante muestreo aleatorio asegurando así la independencia de los datos. El p-value se calculó mediante 9999 simulaciones de Monte Carlo, empleando la corrección de continuidad sugerida por Davison y Hinkley (1997). El resultado obtenido (p-value) es una evidencia moderada en contra de la hipótesis nula de que no hay diferencia entre poblaciones.

Se dispone de un set de datos con información sobre el tiempo que una empresa de telefonía tarda en reparar los problemas que tienen dos grupos de consumidores (CLEC y ILEC). ### Distribución bootstrapping

Mediante bootstrapping, se obtienen nuevas pseudomuestras por separado de cada una de las muestras iniciales manteniendo el tamaño original y se calcula la diferencia.

````python
def calcular_estadistico(x):
    '''
    Función para calcular el estadístico de interés.
    
    Parameters
    ----------
    x : numpy array
         valores de la muestra.
         
    Returns
    -------
    estadístico: float
        valor del estadístico.
    '''
    estadistico = np.mean(x)
    
    return(estadistico)


def bootstraping_2_grupos(x1, x2, fun_estadistico, n_iteraciones=9999):
    '''
    Función para calcular la diferencia de un estadístico entre dos grupos en
    múltiples muestras generadas mediante muestreo repetido con reposición
    (bootstrapping) de forma separada para cada grupo.
    
    Parameters
    ----------
    x1 : numpy array
         valores de la muestra del grupo 1.
         
    x2 : numpy array
         valores de la muestra del grupo 2.
 
    fun_estadistico : function
        función que recibe como argumento una muestra y devuelve el valor
        del estadístico.
        
    n_iteraciones : int
        número iteraciones (default `9999`).
        
    Returns
    -------
    distribuciones: numpy array
        diferencia entre ambos grupos en cada muestra de bootstrapping.
    '''
    
    n1 = len(x1)
    n2 = len(x2)
    dist_boot = np.full(shape=n_iteraciones, fill_value=np.nan)
    
    for i in tqdm(range(n_iteraciones)):
        # Se crea una nueva muestra de cada grupo
        resample_1 = np.random.choice(x1, size=n1, replace=True)
        resample_2 = np.random.choice(x2, size=n2, replace=True)
        # Se calcula el estadístico para cada grupo
        estadistico_1 = fun_estadistico(resample_1)
        estadistico_2 = fun_estadistico(resample_2)
        # Diferencia entre estadísticos
        dist_boot[i]  = estadistico_1 - estadistico_2
        
    return dist_boot
````

````python
dist_boot = bootstraping_2_grupos(
                x1 = datos.Time[datos.Group == 'CLEC'],
                x2 = datos.Time[datos.Group == 'ILEC'],
                fun_estadistico = calcular_estadistico,
                n_iteraciones   = 9999
            )
````

Los datos simulados forman lo que se conoce como bootstrapping distribution y representa la variación esperada en la diferencia de medias debida únicamente al muestreo aleatorio, asumiendo que ambas muestras proceden de dos poblaciones distintas.

### Intervalo basados en percentiles¶

````python
# Intervalo IC basado en percentiles de la distribución bootstrapping
# ==============================================================================
# Un IC del 95% debe abarcar desde el cuantil 0.025 al 0.975
cuantiles = np.quantile(a = dist_boot, q = [0.025, 0.975])
print('-------------------------------')
print('Intervalo basado en percentiles')
print('-------------------------------')
print(cuantiles)
````

````python
# Gráfico intervalo de confianza del 95%
# ==============================================================================
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(7,4))
ax.hist(dist_boot, bins=30, density=True, color='#3182bd', alpha=0.5)
ax.axvline(x=dif_observada, color='red', label='diferencia observada')
ax.axvline(x=cuantiles[0], color='black', linestyle='--', label='IC 95%')
ax.axvline(x=cuantiles[1], color='black', linestyle='--')
ax.hlines(y=0.001, xmin=cuantiles[0], xmax=cuantiles[1], color='black')
ax.set_title('Intervalo bootstrapping basado en percentiles')
ax.set_xlabel('media')
ax.set_ylabel('densidad')
ax.legend();
````

### Conclusión

Sí existen evidencias para afirmar que el tiempo promedio de reparación entre los dos grupos es distinto. El intervalo de confianza del 95% para la diferencia de medias obtenido por bootstrapping indica que, en promedio, las reparaciones del grupo CLEC tardan de 1.7 y 17 horas más.

## Comparación bootstrapping y permutaciones

Tanto los test de permutación como los test de bootstrapping se pueden emplear para estudiar diferencias entre grupos. Existe una lista muy extensa de referencias en las que se debate cuál de los dos métodos es el más adecuado. En general, todas ellas coinciden en que el método más adecuado depende del objetivo de la inferencia, y a su vez, los posibles objetivos están acotados por el diseño de estudio seguido. 

La principal diferencia entre ambos métodos aparece cuando se emplean para calcular p-values. Los test de significancia (cálculo de p-value) se basan en la hipótesis nula de que todas las observaciones proceden de la misma población. El objetivo del test es determinar si la diferencia observada entre los grupos se debe a un determinado factor (tratamiento) o solo a la variabilidad esperada por la naturaleza de un proceso aleatorio.

Cuando la aleatoriedad se debe a la asignación de los sujetos a los distintos grupos, se emplean los test de permutación. La estructura de un experimento que puede analizarse mediante test de permutación es:

* Selección de sujetos del estudio.

* Asignación aleatoria a diferentes grupos

* Aplicación del "tratamiento" y comparación de resultados.

Los test de permutación responden a la pregunta ¿Cuánta variabilidad se espera en un determinado estadístico debido únicamente a la aleatoriedad de las asignaciones, si todos los sujetos proceden realmente de una misma población? Cuando se compara la media entre dos grupos, la pregunta anterior equivale a ¿Qué diferencia entre medias cabe esperar dependiendo de cómo se distribuyan los sujetos en los dos grupos, si todos proceden de una misma población? (Aun siendo todos de una misma población, dado que no serán exactamente idénticos, habrá pequeñas diferencias dependiendo de cómo se agrupen).

Bootstrapping como test de significancia se emplea cuando la aleatoriedad es debida al proceso de obtención de las muestras y no a la asignación en grupos. Responden a la pregunta ¿Cuánta variabilidad se espera en un determinado estadístico debido únicamente al muestreo aleatorio, si todos los sujetos proceden realmente de una misma población? Debido a las pequeñas diferencias entre los individuos de una población, si se extraen dos muestras aleatorias de ella y se comparan, no van a ser exactamente iguales, además, esta diferencia será distinta para cada par de muestras aleatorias extraídas. La estructura de un experimento que puede analizarse mediante bootstrapping es:

* Se obtienen dos muestras aleatorias de dos poblaciones.

* Se comparan.

En conclusión, aunque ambos test pueden emplearse para calcular p-values, sus aplicaciones no se solapan. Los test de permutación y randomization se emplean para diseños experimentales, mientras que el bootstrapping se emplea para diseños muestrales.

Es importante tener en cuenta que ninguno de estos métodos está al margen de los problemas que implica tener muestras pequeñas.
